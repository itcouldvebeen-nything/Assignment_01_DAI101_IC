```python
cat = []        # Categorical columns
num = []        # Numerical columns

# Loop through each column in the dataset
for column in data.columns:
    if pd.api.types.is_numeric_dtype(data[column]):
        if data[column].nunique() > 10:
            num.append(column)  # Add to numerical list if it has more than 10 unique values
        else:
            cat.append(column)  # Otherwise, treat it as categorical
    else:
        cat.append(column)  # Non-numeric columns are treated as categorical

# Print out the categorized lists
print("Categorical columns:", cat)
print("Numerical columns:", num)
```

### Multivarite Analysis

```python
sns.boxplot(x='job_category',y='salary_in_usd',data=data)
plt.xticks(rotation=45, ha='right')
plt.show()
```

> The boxplot shows that "Machine Learning and AI" has the highest median salary, while "Cloud and Database" has the lowest, with significant variation and outliers across all job categories.

```python
data.groupby(by='job_category')['salary_in_usd'].describe()
```

```python
Exp_employmenttype=data.groupby(['work_year','experience_level'])['salary_in_usd'].mean().round(2,).reset_index()
Exp_employmenttype
```

```python
fig=plt.figure(figsize=(8,3))
ax=sns.barplot(data=Exp_employmenttype,x='work_year',y='salary_in_usd',hue='experience_level',errwidth=0,palette = "Blues_d")
for i in ax.containers:
    ax.bar_label(i,color='blue')
plt.show()
```

>The bar chart indicates that executive-level positions consistently earn the highest salaries across all years, with a noticeable peak in 2021, while entry-level positions have the lowest earnings, showing steady growth in salary over the years.

```python
data.groupby('experience_level')['salary'].mean().plot(kind='bar')
print("Average salary for each experience level:")
print(data.groupby('experience_level')['salary'].mean())
plt.show()
```

> This plot shows that emplpoyees with executive level of experience have hoghest average salary

```python
plt.figure(figsize=(26, 10))
sns.scatterplot(x='company_location', y='salary_in_usd', data=data)
plt.xticks(rotation=90,fontsize=6)
plt.xlabel('Company Location')
plt.ylabel('Salary In USD')
plt.title('Relationship Between Salary and Company Location')
plt.show()
```

>The scatter plot shows that the highest salaries are concentrated in locations like the United States and Germany, while many other company locations offer lower salaries with a wide range of values, suggesting significant geographic variation in salary distribution.

```python
sns.violinplot(x='employment_type', y='salary_in_usd', data=data,palette='Set2',legend=False)
plt.xlabel('Employment Type')
plt.ylabel('Salary In USD')
plt.title('Distribution of Salaries Across Employment Types')
plt.show()
```

>The violin plot shows that contract employees have the widest distribution of salaries, while freelance workers generally have the lowest salary range. Full-time and part-time employees fall between these extremes.

```python
sal_byJobcategory=data.sort_values(['salary_in_usd']).groupby(['work_year','job_category'])['salary_in_usd'].mean().reset_index()
sal_byJobcategory
```

```python
fig=plt.figure(figsize=(7,5))
sns.barplot(sal_byJobcategory,x='job_category',y='salary_in_usd',hue='work_year',dodge='bool',palette = "BuGn_r")
plt.xticks(rotation=75)
plt.show()
```

>The bar chart shows that salaries across various data-related job categories generally increased from 2020 to 2023, with Cloud and Database roles seeing a significant spike in 2022. However, Data Analysis and Data Quality & Operations had relatively lower salary growth over these years.
